```python
import pyautogui as pg
import time
```

```python
# define posicoes do mouse

time.sleep(5)
print(pg.position())
```

```python

for i in range(190,201):

    # tira prints
    pg.hotkey("win","shift","s")  # comando windows + shift + s para printar
    time.sleep(1)

    pg.mouseDown(408, 190)  # aperta e segura o clic (esquerdo) na posição
    time.sleep(1)

    pg.moveTo(914, 690, 1)    # move o mouse para a posição final
    time.sleep(0.8)

    pg.mouseUp()   # solta o mouse na posição final


    #abre o paint
    time.sleep(1)
    pg.click(x=651, y=744)

    time.sleep(1)
    pg.click(x=656, y=697)

    # cola a imagem no paint
    time.sleep(0.5)
    pg.hotkey("ctrl","v")

    # salva a imagem no paint
    time.sleep(0.5)
    pg.click(x=45, y=36)     # clica em arquivo

    time.sleep(0.5)
    pg.click(x=75, y=195)    # clica em salvar como

    time.sleep(1)
    pg.moveTo(x=408, y=99)

    time.sleep(1)
    pg.click(x=517, y=129)

    time.sleep(0.5)
    pg.doubleClick(x=513, y=544)

    time.sleep(1)
    pg.hotkey("ctrl","a")

    pg.write(f'{i}')  # salva a imagem com o nome da iteração atual

    time.sleep(0.5)
    pg.press('enter')

    # volta pro site do satélite
    time.sleep(0.7)
    pg.moveTo(x=567, y=745)

    time.sleep(0.5)
    pg.click(x=499, y=649)
   # pg.click(x=546, y=738)

    # avanca para a proxima imagem
    time.sleep(1)
    pg.click(x=256, y=176)  # avança 10 min
    #pg.click(x=302, y=194) # avança 1 hora

    time.sleep(10)
```
